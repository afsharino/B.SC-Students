## Overview
Suppuse we have a chain containing n rings and length of each ring is r. rings 1 and n are installed at a distance of L from each other.

we want to find the cordination of each ring.

The image below shows how each ring i is placed in coordinate plane. We use the Pythagorean theorem to find the coordinates ($x_i, y_i$) of each ring i. in addition to differentiate between left and right rings we suppose that y is negative for left rings and it is positive for right rings.

## Modeling
now we need to modeling our real world problem and then solve the new problem. we know that objects like this chain are positioned so that their gravitational potential energy is minimal.
    
suppose:

$E $ = Gravitational potential energy of system

$E_i$ = Gravitational potential energy of each ring i

so 

$E = \sum{E_i}$ = mg $ \sum (n-i + \frac{1}{2})y_i $

we want this energy be minimum, therefore our objective fuction which we need to minimize it is E.
by solving this problem we can find $y_i$ and by placment $y_i$ in formula $x^2 + y^2 = r^2$ we can find x and in conclution conrdination of each ring.
    
as we mentioned above there are also some restrictions for our problem
1) $\sum y_i = 0$

2) $\sum x_i = L \Rightarrow \sum{\sqrt{(r^2 - y_i^2)}} = L$
        
now the hanging chain problem is modeled as an optimization probelem with equal restrictions:

$Min_y (n-i+\frac{1}{2})y_i$

s.t. $\sum y_i = 0$

$\sum{\sqrt{(r^2 - y_i^2)}}-L = 0$

## Lagrange conditions
    

$\nabla{f(x^*)} + \sum_{i=1}^{m} \lambda_{i}\nabla{g_i(x^*)} = 0_n$

we write this for our problem

$\nabla{f(y)} + \lambda_{1}\nabla{g_1(y)} + \lambda_{2}\nabla{g_2(y)} = 0_n$

$f(y) = \sum{(n-i + \frac{1}{2})} y_i$, 

$g_1(y) = \sum y_i$, 

$g_2(y) = \sum {\sqrt{r^2 - y_i^2} - L}$

$\nabla{f(y)} =  n - i + \frac{1}{2}$,

$\nabla{g_1(y)} =  1$,

$\nabla{g_2(y)} = \frac{-y_i}{\sqrt{r^2 - y_i^2}} $

the system of equations:

\begin{cases}
  n - i + \frac{1}{2} + \lambda_{1} - \lambda_{2}\frac{y_i}{\sqrt{r^2 - y_i^2}} = 0\\
  \sum y_i = 0\\
  \sum {\sqrt{r^2 - y_i^2} - L} = 0\\
  \text{for i from 1 to n}
\end{cases}       

## Calculate $\lambda_1$

$n - i + \frac{1}{2} + \lambda_{1} - \lambda_{2}\frac{-y_i}{\sqrt{r^2 - y_i^2}} = 0$

$\Rightarrow n - i + \frac{1}{2} + \lambda_{1} =  \lambda_{2}\frac{y_i}{\sqrt{r^2 - y_i^2}}$

$\Rightarrow (n - i + \frac{1}{2} + \lambda_{1})^2 =  \lambda_{2}^2\frac{y_i^2}{r^2 - y_i^2}$

$\Rightarrow y_i^2 = \frac{r^2(n - i + \frac{1}{2} + \lambda_{1})^2}{\lambda_{2}^2(n - i + \frac{1}{2} + \lambda_{1})^2} $

$\Rightarrow y_i = \pm \frac{r(n - i + \frac{1}{2} + \lambda_{1})}{\sqrt{\lambda_{2}^2(n - i + \frac{1}{2} + \lambda_{1})^2}} $

Now we need to detect which sign is correct. We know that lef-side rings and right-side rings are symetric and their signs are opposite so the following relation is established.

in $y_i = \pm \frac{r(n - i + \frac{1}{2} + \lambda_{1})}{\sqrt{\lambda_{2}^2(n - i + \frac{1}{2} + \lambda_{1})^2}} $ formula r in the numerator is constant so it has no effect on the sign. same condition for Radical in denominator, so the only thing we need to check is $n - i + \frac{1}{2} + \lambda_{1}$.


$n - i + \frac{1}{2} + \lambda_{1} = - [n - (n-i+1) + \frac{1}{2} + \lambda_{1}] $ $\forall i \in [1, n]$

$\Rightarrow n - i + \frac{1}{2} + \lambda_{1} = - [n-n+i-1 + \frac{1}{2} + \lambda_{1}] $

$\Rightarrow n - i + \frac{1}{2} + \lambda_{1} = -i+ \frac{1}{2} - \lambda_{1} $

$\Rightarrow n = -2 \lambda_{1} $

$\Rightarrow \lambda_{1} = \frac{-n}{2} $

lets inplace $\lambda_{1}$ in equation.

$y_i = \pm \frac{r(\frac{n}{2} - i + \frac{1}{2})}{\sqrt{\lambda_{2}^2(\frac{n}{2} - i + \frac{1}{2})^2}} $

simplify by multiplying to 2

$y_i = \pm \frac{r(n - 2i + 1)}{\sqrt{4\lambda_{2}^2(n - 2i + 1)^2}} $

Which sign?

as we know that for left-side rings $y_i$ < 0 so 

$y_1$ < 0 (*)

therfore we calculate the equation for i = 1 to see the result.

$y_1 = \pm \frac{r(n - 1)}{\sqrt{4\lambda_{2}^2(n - 1)^2}} $

as we know r > 0 and n-1 = 0 so numerator is + and denominator is + too. (**)

from (*) and (**) we conclude that $y_i = - \frac{r(n - 2i + 1)}{\sqrt{4\lambda_{2}^2+(n - 2i + 1)^2}} $

## Calculate $\lambda_2$

We use the 2nd restriction of the problem to find lambda

$\sum {\sqrt{r^2 - y_i^2}} = L$

by inplacing $y_i$ from previous part we have

$\sum {\sqrt{r^2 - \frac{r^2(n - 2i + 1)^2}{4\lambda_{2}^2(n - 2i + 1)^2}}} = L$

Now by simplifying the above equation we have

$\sum {\sqrt{\frac{4r^2 \lambda_{2}^2 + r^2 (n - 2i + 1)^2 - r^2(n - 2i + 1)^2 }{4\lambda_{2}^2(n - 2i + 1)^2}}} = L$

$\Rightarrow \sum{\frac{\pm 2r \lambda_{2}} {\sqrt{4\lambda_{2}^2(n - 2i + 1)^2}}} = L$

Now we need to detect wich sign is correct. to do this we use $n - i + \frac{1}{2} + \lambda_{1} - \lambda_{2}\frac{y_i}{\sqrt{r^2 - y_i^2}} = 0$ formula for i = 1 and $\lambda_1 = \frac{-n}{2}$.

$\frac{n}{2} - 1 + \frac{1}{2} - \lambda_{2}\frac{y_1}{\sqrt{r^2 - y_1^2}}$

$\Rightarrow \frac{n-1}{2} = \lambda_{2}\frac{y_1}{\sqrt{r^2 - y_1^2}}$

$\Rightarrow \lambda_{2} = (\frac{n-1}{2})\frac{\sqrt{r^2 - y_1^2}}{y_1}$

as we know $y_1$ < 0 and n-1 > 0 so $\lambda_{2}$ < 0

now lets back to our formula, as r > 0 and $\lambda_{2}$ < 0 therfore we have

$-\sum{\frac{ 2r \lambda_{2}} {\sqrt{4\lambda_{2}^2(n - 2i + 1)^2}}} = L$

$\Rightarrow L + 2r \lambda_{2} = \sum{\frac{ 1} {\sqrt{4\lambda_{2}^2(n - 2i + 1)^2}}} =0$

The Algorithm should get n,r and L as input and return $\lambda_1$, $\lambda_2$,$y_1, ..., y_n$ as an output.

```python
# Get input from the user.
n = int(input("Enter n: "))
r = int(input("Enter r: "))
L = int(input("Enter L: "))
```

```python
def find_lambda1(n):
    return -(n/2)
```

```python
import mpmath
from sympy import Symbol, nsolve, sqrt
from sympy.concrete.summations import Sum

def find_lambda2(n, L, r):
    lambda2 = Symbol('l')
    i = Symbol('i')

    summation = Sum(1 / sqrt( (4*(lambda2**2)) + (n-(2*i)+1)**2), (i, 1, n))

    f = L - 2*r*lambda2*summation
    return (nsolve(f, 0))
```

```python
def find_y_i(n, r, lambda2):
    list_of_y = []
    for i in range(1, n+1):
        y_i = -((r*(n-2*i+1)) / sqrt(4*lambda2**2 + (n-2*i+1)**2))
        list_of_y.append(y_i)
        
    return list_of_y
```

```python
list_of_y = find_y_i(n, r, find_lambda2(n,L, r))

def find_x_i(list_of_y, r):
    list_of_x = []
    
    for y in list_of_y:
        x_i = sqrt(r**2 - y**2)
        list_of_x.append(x_i)
        
    return list_of_x

list_of_x = find_x_i(list_of_y, r)
```
