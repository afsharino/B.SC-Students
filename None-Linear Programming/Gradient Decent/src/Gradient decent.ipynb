```python
import sympy as sp

# Create two symbols p and q.
p, q = sp.symbols("p q")

# f is our function.
f = p**4 - 2*p*q + 2*q**2 + -p+q+3

# Display the result after differentiate
print(f"                                                 [ {sp.diff(f, p)} ]")
print("The gradiant of '2*p**2 + 2*p*q + q**2 + p-q' is:|               |")
print(f"                                                 [ {sp.diff(f, q)} ]")
```

<div class="alert alert-block alert-success">
    <h1 align="center">Main Algorithm</h1><br\>
</div>

```python
# This modjule is for calculating gradiant
import numdifftools as nd 

import numpy as np

import math
```

```python
# This is our function
f = lambda x:2*x[0]**2 + 2*x[0]*x[1] + x[1]**2 + x[0]-x[1]

# 1D-array to store x_i vectores.
# array is initialized with zeos
x = np.zeros((101,2))

# Initialize primary point 
x[0] = [1, 1]

# Step length
eta = 0.1

epsilon = 0.001
```

```python
for k in range(0, 100):
    
    # Calculate the gardient
    gradiant = nd.Gradient(f)([x[k][0], x[k][1]])
    
    # Formula
    x[k+1] = x[k] - eta * gradiant
    
    # Display x_k and its value
    print("_______________________________")
    print(f"iteration number {k+1}: ")
    print(f"x_{k} = {x[k]}")
    print(f"f(x_{k}) = {f(x[k])}")
    print("_______________________________")
    print(f"               {chr(8595)}        ")
    print(f"               {chr(8595)}        ")
    print(f"               {chr(8595)}        ")
    
    # stopp criterion: norm(x[k+1]-x[k]) / norm(x[k]) < epsilon
    if np.linalg.norm(x[k+1]-x[k]) / np.linalg.norm(x[k]) < epsilon:
        break

# Display minimum poin and its value
xstar = [-1, 3/2]
fstar = f(xstar)
print(f"        x* is {xstar}.")
print(f"        f(x*) is {fstar}.")

vectores = x.copy()
```
